```python
import pandas as pd
import numpy as np
import edhec_risk_kit_206 as erk
```

```python
ind_rets = erk.get_ind_returns(weighting="vw",n_inds=49)["2014":"2018"]
ind_mcap = erk.get_ind_market_caps(n_inds=49,weights=True)["2014":"2018"]
cov = ind_rets.cov() * 12
cap_weights = ind_mcap.iloc[0]
```

```python
#cap weighted
erk.risk_contribution(w=cap_weights,cov=cov).idxmax()   
```

```python
#cap weighted
erk.risk_contribution(w=cap_weights,cov=cov).max()
```

```python
#equally weighted
eq_weights = pd.Series(1/49, index=cov.index)
erk.risk_contribution(w=eq_weights,cov=cov).idxmax()
```

```python
erk.risk_contribution(w=eq_weights,cov=cov).max()
```
